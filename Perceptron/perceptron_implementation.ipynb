# Basic Perceptron Demo

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student_exam_data_new.csv")
df
```

```python
df.shape
```

```python
sns.scatterplot(data =df,x="Study Hours",y="Previous Exam Score",hue= "Pass/Fail")
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
```

```python
p.fit(X,y)
```

```python
p.coef_
```

```python
p.intercept_
```
