The sequential API allows you to create models layer-by-layer for most problems. It is limited in that it does not allow you to create models that share layers or have multiple inputs or outputs.

# Functional API
- The functional API in Keras is an alternate way of creating models that offers a lot more flexibility, including creating more complex models.
- with the help of functional api, we can build non linear neural networks.
- used to make non-linear topology


```python
from tensorflow.keras.layers import *
from tensorflow import keras
from tensorflow.keras.models import Model
from tensorflow.keras.utils import plot_model
```

# Build Model with One Input and one Output

```python
x = Input(shape=(3,3))
hidden1 = Dense(128,activation='relu')(x)
hidden2 = Dense(64,activation = 'relu')(hidden1)
output1  = Dense(1,activation = 'sigmoid')(hidden2)
```

```python
model = Model(inputs=x,outputs =[ output1])
```

```python
plot_model(model,show_shapes=True)
```

# Build Model with two input and two output

```python
inputA = Input(shape=(3,3))
inputB = Input(shape=(3,3))

vijay = concatenate([inputA,inputB])
hidden1 = Dense(128,activation='relu')(vijay)


output1 = Dense(1,activation='sigmoid')(hidden1)
output2 = Dense(1,activation='sigmoid')(hidden1)
```

```python
model2= Model(inputs = [inputA,inputB],outputs =[output1,output2])
plot_model(model2)
```

# Shared Layer Model

```python
InputA = Input(shape=(64,64,1))

hidden1 = Conv2D(32,kernel_size =(2,2) ,activation='relu')(InputA)
Pooling1 = MaxPooling2D(pool_size=(2,2),strides=2)(hidden1)
flatten1 = Flatten()(Pooling1)


hidden2 = Conv2D(16,kernel_size= (2,2),activation='relu')(InputA)
Pooling2 = MaxPooling2D(pool_size=(2,2),strides=2)(hidden2)
flatten2 = Flatten()(Pooling2)

concatenate1 = concatenate([flatten1,flatten2])

dense1 = Dense(32,activation='relu')(concatenate1)
output = Dense(1,activation='sigmoid')(dense1)
```

```python
model3 = Model(inputs=InputA,outputs=output)
plot_model(model3)
```
