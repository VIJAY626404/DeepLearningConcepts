# Dropout in Neural Network:
- randomly drop node from input layer and hidder layer that means traning same data with different different nerual network.
- to handle overfitting problem dropout play crucial rule that help to improve the validation accuracy of neural network.
- if model is overfitted then increase the value dropout ratio, if model is underfitted then decrease the value of dropout ration. 
- node will drop only training time , in testing time every node will present. but we can't assign total weight .
- weights of node in testing time = W(1-p), where W = intial weight of node,1-p = probility of not dropout node in neural network

# Some tips to use dropout:
- if neural network is overfitted the increase the ration of dropout
- if neural network is underfitted decrease the ratio of dropout 
- range of dropout value for different models:
    -- for ANN it's lie between 10 % - 50 %
    -- for CNN it's lie between 40 - 50 %
    -- for RNN it's lie between 20 % - 30 %

# Drawback of Dropout:
- convergence delay due to slow traing
- value of loss function / cost function will change due to dropout so , calculation of gradient will difficult
